```python
import numpy as np

import pandas as pd
```

```python
def my_df( ):
    
    myarray = np.array([ [ 'xwc' , '2020-01-01 14:00:00' , 1 , 1 , 3 , 'v' , 'a' , 0 ] , 
                         [ 'swd' , '2020-01-02 08:00:00' , 2 , 2 , 2 , 'v' , 'b' , 1 ] ,
                         [ 'klp' , '2020-01-03 23:00:00' , 4 , 5 , 1 , 'w' , 'c' , 1 ] ,
                         [ 'rft' , '2020-01-04 13:00:00' , 2 , 1 , 9 , 'w' , 'c' , 1 ] , 
                         [ 'azp' , '2020-01-05 12:00:00' , 1 , 0 , 7 , 'w' , 'b' , 1 ] ] )

    col = [ 'id' , 'col_dat_str' , 'col_num_01', 'col_num_02', 'col_num_03' , 'col_cat_01' , 'col_cat_02' , 'target' ]

    df = pd.DataFrame( myarray , columns = col )

    df.set_index( 'id' , drop = True , append = False , inplace = True , verify_integrity = False )


    df[ 'col_num_01' ] = df[ 'col_num_01' ].astype( 'float64' )

    df[ 'col_num_02' ] = df[ 'col_num_02' ].astype( 'float64' )

    df[ 'col_num_03' ] = df[ 'col_num_03' ].astype( 'float64' )

    df[ 'target' ] = df[ 'target' ].astype( 'float64' )

    return df
```

```python
df = my_df()

df.head( 5 )
```

```python
df.dtypes
```

# Trick 01 : Fix messy dataframe column names

```python
df = df.rename( columns = { 'col_num_01' : 'col num $01' } )

df.columns
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace( ' ' , '_' ).str.replace( '(' , '' ).str.replace( ')' , '' ).str.replace( '$' , '' )

df.columns
```

```python
df.head( 5 )
```

# Trick 02 : Remove Outliers

```python
df.loc[ df[ 'col_num_01' ] < np.percentile( df[ 'col_num_01' ] , 99 ) ]
```

# Trick 03 : Calculate Percentage Change

```python
df[ 'perc_change' ] = df[ 'col_num_03' ].pct_change()
```

```python
df.head(  5 )
```

# Trick 04 : Left merge with indicator

```python
df2 = my_df()

df2.head( 5 )
```

```python
pd.merge( df , df2 , how = 'outer' , on = 'id' , suffixes = ( '_left' , 'right' ) , indicator = True )
```

# Trick 05 : Compress & Save Pandas Dataframe

```python
df.to_csv( 'df.gz', compression = 'gzip' , index = False )
```

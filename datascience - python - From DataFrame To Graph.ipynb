```python
import pandas as pd
```

```python
df = pd.read_csv( 'https://raw.githubusercontent.com/ecdedios/networkx-quick-start/master/jira_sample.csv' )
```

```python
df.head()
```

```python
df = df[ [ 'Assignee' , 'Reporter' ] ]
```

```python
df.head()
```

```python
import networkx as nx
```

```python
G = nx.from_pandas_edgelist( df , 'Assignee' , 'Reporter' )
```

```python
from matplotlib.pyplot import figure
```

```python
figure( figsize = ( 10 , 8 ) )

nx.draw_shell( G , with_labels = True )
```

# DataFrame with Nodes and Number of Connexions

```python
leaderboard = { }


for x in G.nodes :
   leaderboard[ x ] = len( G[ x ] )
   
s = pd.Series( leaderboard , name = 'connections' )

df_cnx = s.to_frame().sort_values( 'connections' , ascending = False )

df_cnx.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
import pickle
```

```python
import pandas
```

```python
# generate 2d classification dataset
X, y = make_blobs(n_samples=100, centers=2, n_features=4, random_state=1)
```

```python
# fit final model
model = LogisticRegression()
model.fit(X, y)
```

```python
###### Save and Load Machine Learning Models in Python with scikit-learn
```

```python
# save the model to disk
filename = 'finalized_model_logistic_regression.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
# New dataset
X_new, _ = make_blobs(n_samples=150, centers=2, n_features=4, random_state=1)
```

```python
# load the model from disk

loaded_model = pickle.load( open( filename, 'rb') )

# Make class prediction y_new on X_new

y_new = loaded_model.predict(X_new)
```

```python
# Probabilit√© d'appartenance aux classes ( 0 ou 1 )

y_new_prob = loaded_model.predict_proba(X_new)
```

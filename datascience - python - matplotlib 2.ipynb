```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv( 'https://coded2.herokuapp.com/datavizpandas/london2018.csv' )
```

```python
weather
```

# Line Graph

```python
weather.plot( x = 'Month' ,
              y = 'Tmax' )
```

```python
weather.plot( x = 'Month',
              y = [ 'Tmax' , 'Tmin' ] )
```

# Bar Charts

```python
weather.plot.bar( x = 'Month' , 
                  y = 'Tmax' )
```

```python
weather.plot.barh( x = 'Month' , 
                   y = 'Tmax' )
```

```python
weather.plot.bar( x = 'Month' , 
                  y = [ 'Tmax' , 'Tmin' ] )
```

# Multiple charts

```python
weather.plot.bar( x = 'Month' , 
                  y = [ 'Tmax' , 'Tmin' , 'Rain' , 'Sun' ] ,
                  subplots = True ,
                  layout = ( 2 , 2 ) )
```

# Scatter Plot

```python
weather.plot.scatter( x = 'Sun' ,
                      y = 'Rain' )
```

# Saving the Charts

```python
weather.plot( x = 'Month' , 
              y = [ 'Tmax' , 'Tmin' , 'Rain' , 'Sun' ] ,
              subplots = True ,
              layout = ( 2 , 2 ) )
```

```python
plt.savefig( 'line_charts.jpg' )
```

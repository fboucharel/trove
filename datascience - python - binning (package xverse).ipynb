## 0 - Packages

```python
import pandas as pd
```

```python
from xverse.feature_subset import FeatureSubset

from xverse.feature_subset import SplitXY

from xverse.transformer import MonotonicBinning

from xverse.transformer import WOE

from xverse.ensemble import VotingSelector
```

## 0 - Fonction

```python
def create_Xy( ) :
    
    df = pd.read_csv( 'https://raw.githubusercontent.com/Sundar0989/XuniVerse/master/data/bank.csv'  , sep = '|' , header = 0 , error_bad_lines=False )
    
    # df avec uniquement colonnes numÃ©riques :
    num_col = list( df._get_numeric_data().columns ) 
    fsub = FeatureSubset( num_col )
    df = fsub.fit_transform( df )
    
    # df -> X , y :
    Xy = SplitXY( [ 'target' ] )
    X , y = Xy.fit_transform( df )
    
    return X , y
```

## I - Monotonic binning

```python
X , y = create_Xy()
```

```python
mono_bng = MonotonicBinning()

mono_bng.fit( X , y )
```

```python
out_bins = mono_bng.bins

print( out_bins )
```

```python
X_mono_bng = mono_bng.transform( X )

X_mono_bng.head()
```

```python
X_new , y_new = create_Xy()

mono_bng = MonotonicBinning( out_bins )

mono_bng.fit( X , y )

X_new_tsf = mono_bng.transform( X_new )

X_new_tsf.head()
```

## II - Woe

```python
X , y = create_Xy()
```

```python
woe = WOE( woe_prefix = 'woe' )

woe.fit( X , y )
```

```python
woe.woe_df.head(10)
```

```python
woe.iv_df
```

```python
out_woe_bins = woe.woe_bins

print( out_woe_bins )
```

```python
out_mono_bins = woe.mono_custom_binning

print( out_mono_bins )
```

```python
X_woe_tsf = woe.transform( X )

X_woe_tsf.head()
```

```python
X_new , y_new = create_Xy()

woe = WOE( woe_bins = out_woe_bins , mono_custom_binning = out_mono_bins )

X_new_tsf = woe.transform( X_new )

X_new_tsf.head()
```
